```python
## Importing NLTK (Natural Language Toolkit)
import nltk
```

```python
## Importing Regular Expression
import re
```

```python
## Downloading Soppwords from NLTK library
nltk.download("stopwords")
```

```python
## Importing stopwords
from nltk.corpus import stopwords
```

```python
## Importing PorterStemmer
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
## Checking all Stopworks present in NLTK 
stopwords.words('english')
```

```python
##Checking Sub-string
rev =re.sub('[^a-zA-Z]', ' ',data['message'][0])
```

```python
rev
```

```python
## Ching all sub-string into lower case
rev.lower()
```

```python
## Splitting every sub-string
rev.split()
```

```python
## Seperating every sub-string which is not present in stopwords
rev = [ps.stem(word) for word in rev if not word in stopwords.words('english')]
```

```python
rev
```

```python
## creating black list as corpus
corpus = []
```

```python
## doing for loop and extracting all sentences which ignore stopwords and return in lower case.
for i in range(0,len(data)):
    review= re.sub('[^a-zA-Z]', ' ',data['message'][i])
    review= review.lower()
    review= review.split()
    
    review= [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '. join(review)
    corpus.append(review)
    
    
    
    
```

```python
## Returning 1st sentences which are without Stopwords present in lower cases
corpus[0]
```

```python
## Returning all sentences which are without Stopwords present in lower cases
corpus
```

```python
## Importing CountVectorizer from sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 2500)
```

```python
## Fitting the model
X = cv.fit_transform(corpus).toarray()
```

```python
#1 set of unique words
#2 finally it is creating a vector

X.shape
```

```python
X[0]
```

```python
len(X[0])
```

```python
data['label']
```

```python
y= pd.get_dummies(data['label'],drop_first=True)
```

```python
X
```

```python
y
```

## Splitting Train_Test_

```python
## Splitting Train_Test_
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

## Building Model using GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
## Fitting a GaussianNB Model.
model.fit(X_train,y_train)
```

```python
## Predicting the model
y_pred=model.predict(X_test)
```

```python
## Importing accuracy_score from sklearn
from sklearn.metrics import accuracy_score
```

```python
## Checkig Accuracy of the Model
accuracy_score(y_test,y_pred)
```

## Building Model using MultinomialNB 

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model2=MultinomialNB()
```

```python
## Fitting the MultinomialNB Model
model2.fit(X_train,y_train)
```
