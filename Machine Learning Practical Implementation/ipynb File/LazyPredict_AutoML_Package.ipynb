## **What is LazyPredict in Machine Learning ?**

* lazypredict is a convenient wrapper library, that enables us to quickly fit all the models to our dataset and compare their performance. This way, we can see what works well “out of the box”.

### <span style = 'color : blue' >   Importing All Necessary Library   </span>

```python
from lazypredict.Supervised import LazyClassifier, LazyRegressor
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

# <span style = 'color : red' >   Classification Model task   </span>

```python
# load data
data = datasets.load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

# fit all models
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

### <span style = 'color : blue' >   Checking All Classification Model Accuracy in details:-   </span>

```python
models
```

### <span style = 'color : blue' >   Checking All Classification Model Prediction in details:-   </span>

```python
predictions
```

# <span style = 'color : Red' >   Regression Model task   </span>

```python
# load data
boston = datasets.load_boston()
X, y = boston.data, boston.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

# fit all models
reg = LazyRegressor(predictions=True)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
```

### <span style = 'color : blue' >   Checking All Regressor Model Accuracy in details:-   </span>

```python
models
```

### <span style = 'color : blue' >    Checking All Regressor Model Prediction in details:-   </span>

```python
predictions
```
